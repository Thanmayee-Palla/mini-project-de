```sql

USE plworkforce_catalog.003_gold;

-- DIM COMPANY (descriptive)
CREATE OR REPLACE TABLE dim_company AS
SELECT
    company_id,
    company_name,
    industry,
    country
FROM plworkforce_catalog.002_silver.company;

-- DIM DEPARTMENT (descriptive)
CREATE OR REPLACE TABLE dim_department AS
SELECT
    department_id,
    department_name
FROM plworkforce_catalog.002_silver.departments;

-- DIM EMPLOYEE (descriptive)
CREATE OR REPLACE TABLE dim_employee AS
SELECT
    employee_id,
    full_name,
    department_id,
    hire_date,
    position
FROM plworkforce_catalog.002_silver.employee;

-- DIM ACCOUNT (descriptive)
CREATE OR REPLACE TABLE dim_account AS
SELECT
    account_id,
    account_type,
    is_active
FROM plworkforce_catalog.002_silver.accounts;

-- Verify dimension tables created
SELECT 'dim_company' AS table_name, COUNT(*) AS row_count FROM dim_company
UNION ALL
SELECT 'dim_department' AS table_name, COUNT(*) AS row_count FROM dim_department
UNION ALL
SELECT 'dim_employee' AS table_name, COUNT(*) AS row_count FROM dim_employee
UNION ALL
SELECT 'dim_account' AS table_name, COUNT(*) AS row_count FROM dim_account;
```
