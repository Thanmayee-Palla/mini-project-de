```python
 
files = [
    'employee.csv',
    'departments.csv',
    'company.csv',
    'general_ledgers.csv',
    'payroll.csv',
    'accounts.csv'
]
 
 
s3_base_path = 	's3://de-mini-project-123/'
 
for file in files:
    table_name = file.replace('.csv', '')
   
    df = spark.read.csv(s3_base_path + file, header=True, inferSchema=True)
   
    target_table = f"plworkforce_catalog.001_bronze.{table_name}"
    df.write.mode("overwrite").saveAsTable(target_table)
   
 
print("\n All files loaded to bronze schema successfully!")
 
```
